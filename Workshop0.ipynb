Task- 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).

2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).

4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.
• Conversion Options:
1. Length:– Convert meters (m) to feet (ft).– Convert feet (ft) to meters (m).
2. Weight:– Convert kilograms (kg) to pounds (lbs).– Convert pounds (lbs) to kilograms (kg).
3. Volume:– Convert liters (L) to gallons (gal).– Convert gallons (gal) to liters (L)

```python
def convert_units(conversion_type, value):
    """
    Converts units based on the conversion type.

    Parameters:
    conversion_type (str): Type of conversion selected by the user.
    value (float): Numeric value to convert.

    Returns:
    float: Converted value.
    """

    if conversion_type == "1":  # meters to feet
        return value * 3.28084
    elif conversion_type == "2":  # feet to meters
        return value / 3.28084
    elif conversion_type == "3":  # kg to lbs
        return value * 2.20462
    elif conversion_type == "4":  # lbs to kg
        return value / 2.20462
    elif conversion_type == "5":  # liters to gallons
        return value * 0.264172
    elif conversion_type == "6":  # gallons to liters
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    choice = input("Choose conversion type (1-6): ")

    try:
        value = float(input("Enter value to convert: "))
        result = convert_units(choice, value)

        if result is not None:
            print("Converted value:", result)
        else:
            print("Invalid conversion type selected.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")


if __name__ == "__main__":
    main()
```

1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
• Requirements:– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list
```

```python
nested = [[1, 2], [3, 4], [5]]

result = flatten(nested)

print("Original nested list:", nested)
print("Flattened list:", result)
```

the elements.– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call for nested list
        else:
            total += item
    return total
```

```python
def remove_element(lst, elem):
    new_list = []
    for sublist in lst:
        # Only process if the item is a list
        if isinstance(sublist, list):
            filtered = [x for x in sublist if x != elem]
            new_list.append(filtered)
        else:
            # If the top-level item is not a list, skip or remove if it matches elem
            if sublist != elem:
                new_list.append(sublist)
    return new_list
```

5. Find the Maximum Element in a Nested List:
Write a Python function that finds the maximum element in a nested list (regardless of depth).
• Requirements:– Define a function find
element.
max(lst) that takes a nested list lst and returns the maximum– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_element_max(lst):
    max_val = float('-inf')  # Start with the smallest possible value
    for item in lst:
        if isinstance(item, list):
            # Recursively find max in nested list
            nested_max = find_element_max(item)
            if nested_max > max_val:
                max_val = nested_max
        else:
            if item > max_val:
                max_val = item
    return max_val
```

```python
nested = [[1, 2], [3, [4, 5]], 6]

result = find_element_max(nested)

print("Nested list:", nested)
print("Maximum element:", result)
```

6. Count Occurrences of an Element in a Nested List:
Write a Python function that counts how many times a specific element appears in a nested list.

occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.– Example: Fortheinput lst = [[1, 2], [2, 3], [2, 4]]andelem = 2,theoutputshould
be 3

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recursively count in nested list
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count
```

7. Flatten a List of Lists of Lists:
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.
• Requirements:– Define a function deep
flatten(lst) that takes a deeply nested list lst and returns a single
f
lattened list.– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursively flatten nested lists
        else:
            flat_list.append(item)
    return flat_list
```

```python
nested = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

result = deep_flatten(nested)

print("Original deeply nested list:", nested)
print("Flattened list:", result)
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.
• Requirements:– Define a function average
of all the elements.
nested(lst) that takes a nested list lst and returns the average– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5

```python
def average_nested(lst):
    def helper(sublist):
        total = 0
        count = 0
        for item in sublist:
            if isinstance(item, list):
                t, c = helper(item)  # recursively get sum and count
                total += t
                count += c
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = helper(lst)
    if total_count == 0:
        return 0  # Avoid division by zero
    return total_sum / total_count
```

```python
nested = [[1, 2], [3, 4], [5, 6]]

result = average_nested(nested)

print("Nested list:", nested)
print("Average of all elements:", result)
```

3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1,000,000. Measure and

Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.

```python
# Create lists and arrays
N = 1_000_000

# Python lists
list1 = list(range(N))
list2 = list(range(N))

# NumPy arrays
array1 = np.arange(N)
array2 = np.arange(N)
```

```python
# Dot product using Python lists
start_time = time.time()

# Using sum of element-wise multiplication
dot_list = sum(list1[i] * list2[i] for i in range(N))

end_time = time.time()
print("Time taken using Python lists (dot product): {:.6f} seconds".format(end_time - start_time))
```

```python
# Dot product using NumPy arrays
start_time = time.time()

# NumPy dot product
dot_array = np.dot(array1, array2)

end_time = time.time()
print("Time taken using NumPy arrays (dot product): {:.6f} seconds".format(end_time - start_time))
```

4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Mea
sure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
# Create matrices
N = 1000

# Python lists (nested lists)
list1 = [[i + j for j in range(N)] for i in range(N)]
list2 = [[i - j for j in range(N)] for i in range(N)]

# NumPy arrays
array1 = np.array(list1)
array2 = np.array(list2)
```

```python
# Matrix multiplication using Python lists
start_time = time.time()

# Initialize result matrix
result_list = [[0 for _ in range(N)] for _ in range(N)]

# Naive nested loop multiplication
for i in range(N):
    for j in range(N):
        for k in range(N):
            result_list[i][j] += list1[i][k] * list2[k][j]

end_time = time.time()
print("Time taken using Python lists (matrix multiplication): {:.6f} seconds".format(end_time - start_time))
```

```python
#Matrix multiplication using NumPy arrays
start_time = time.time()

# NumPy matrix multiplication (vectorized)
result_array = np.dot(array1, array2)

end_time = time.time()
print("Time taken using NumPy arrays (matrix multiplication): {:.6f} seconds".format(end_time - start_time))
```
